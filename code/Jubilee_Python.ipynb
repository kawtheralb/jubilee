```python
import pandas as pd

# Specify the path to my CSV file
file_path = '/content/drive/My Drive/Spring 2024/Digital Traces/rawdata/allcomments.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
from pysentimiento import create_analyzer
```

```python
# Create sentiment analyzer
sentiment_analyzer = create_analyzer(task="sentiment", lang="en")

# Create emotion analyzer
emotion_analyzer = create_analyzer(task="emotion", lang="en")

# Create hate speech analyzer
hate_analyzer = create_analyzer(task="hate_speech", lang="en")
```

```python
sentiment_scores = []
emotion_predictions = []
hate_predictions = []

for comment in df['text']:
    sentiment_output = sentiment_analyzer.predict(comment)
    emotion_output = emotion_analyzer.predict(comment)
    hate_output = hate_analyzer.predict(comment)


    sentiment_scores.append(sentiment_output.probas)
    emotion_predictions.append(emotion_output.probas)
    hate_predictions.append(hate_output.probas)

# After the loop, you can assign these lists to DataFrame columns
df['sentiment_scores'] = sentiment_scores
df['emotion_predictions'] = emotion_predictions
df['hate_predictions'] = hate_predictions
```
